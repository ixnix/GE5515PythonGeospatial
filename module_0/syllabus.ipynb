## GE5515 Python for Geospatial Analysis

### Overview
Python is a versatile, high-level, interpreted programming language and one of the most widely used open-source platforms for Earth science data analysis. Its popularity in Earth and environmental sciences stems from its user-friendly syntax and an extensive ecosystem of tools that support data processing, analytics, and computations. This course introduces Python as a powerful tool for geospatial analysis, guiding students through key concepts and practical applications. Students will learn to work with geospatial packages, manage spatial data formats, perform geospatial analyses, and create maps. The course also covers advanced topics like geocoding, spatial joins, overlay analysis, and raster and terrain analysis. By the end of the course, students will have the practical expertise to apply Python in diverse geospatial contexts, enabling them to analyze, visualize, and share geospatial insights.

### Course Format
This course is organized as weekly learning modules. Each module consists of two meetings: one for lecture and code demonstration, and the other dedicated for computational labs. During the labs, students will work on homework assignments and coding problems in a collaborative and interactive environment. Lecture notes and computational assignments are provided as Jupyter Notebooks.

### Student Learning Outcome
By the end of this course students are expected to be able to:
- Utilize Python libraries to manage, analyze, and visualize spatial and non-spatial data across various formats.
- Create and manipulate geometric objects, manage coordinate reference systems, and perform spatial data transformations.
- Conduct geospatial analyses such as geocoding, spatial joins, overlay, and proximity-based queries.
- Process and analyze raster data, including reclassification, mosaic, and spatial calculations for advanced geospatial workflows.
- Design and share static and interactive maps, leveraging Python tools for effective visualization and communications.

The final grade will be evaluated based on three components:

- **Computational Labs (50%)**. Computational labs are designed to test learning of each module. Students should carefully follow the instructions and type answers in designated cells (in markdown or live codes). Each assignment has a full credit of 100 points. 5 points will be deducted for each incorrect answer or failure to meet specific requirements, e.g., if you use [hard-coding](https://www.reddit.com/r/learnpython/comments/vznhh1/what_does_hardcoded_means/?rdt=40086) or reserved words as variable names. - **Class Project (40%)**. Students will work on a self-directed class project, where they identify a geospatial problem of their choice. This project will consist of collecting, preprocessing, and analyzing spatial and non-spatial data using Python. Students will design and implement a comprehensive workflow, leveraging the tools and techniques covered in the course to address their chosen problem. The final deliverable will be a **class presentation** and a **project report** that outlines the project's objectives, data sources, technical approach, analytical workflow, and key findings. The project report <u>must be written in the format of Jupyter Notebook</u>. All data files used during the project must be submitted for verification.
